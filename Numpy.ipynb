#Deep and Shallow Copy

```python
## DEEP COPY
##> To understand the concept of deep copy, let us consider an example :
# Let us create a vector :
v3=np.arange(1,10)
print(f'{v3}')

# And now let us silce the this vector
s = v3[:6]
print(s+10)

##> As you can see that we have done some changes in the vector .
#   Does that affect the original vector . Let's check it by printing the numpy vector in the next cell .
```

```python
print(v3)

##> As you can see that the numpy vector has not being changed at all.
#  This is the concept of Shallow copy , it means that the silced part has created a seprate memeory space in the RAM (Random Acess Memory).
#  It means that the changes made in this part would not affect the original numpy vector .
#  Because the silce variable is not targeting the memeory location on the numpy vector .
```

```python
# Deep copy
##> Opposite to the shallow copy , means if we make changes in the numpy array  in any way it would be reflected in the original numpy vector .
#   Like for example:
v4=np.arange(20,30)
print(f'{v4}')

# Now lets us keep is v4 vector in some other variable .
V = v4

# And now if we make cjhanges in the vector :

V[5]=56
print(f'\n{V}\n')

# Here as you can see that there is change in the element in the vector .
# Does that meant there is a change in the original vector too ?
print(f'\n{v4}\n')

##> As you can clearly see that there are changes in the original vector .
##> And this is the concept on deep copy .
#   In deep copy the memory address or location is of one varible is targeted by both variable ,
#   that mean if we change anything in one variable that change will apper in    another variable.
```

```python
##? But if we want to store a numpy array or vector in some another variable    and we dont want to be changed .
#   Then we can simply create it's copy . Like :
V1=v4.copy()
print(f'{v4}')


# And if we change something :
V1[6] = 99
print(f'\n{V1}\n')
print(f'{v4}\n')

##> As you can see that now if change anything There is no effect in original numpy array or vector
```

#Matrix operations

```python
mat1=np.arange(1,5).reshape(2,2)
mat2=np.arange(5,9).reshape(2,2)

# Now our first opeartion is , matrix multiplication

''''''
# 1.Dot product

##> As we have seen above some of arithmetic operation that we can multiply the matrices or vectors simply putting a multiplication symbol in between them .
#   But that multiplication was index wise multiplication not actual matrix multiplication
#   So for that we can do :

multiply1= mat1 @ mat2
print(f'Matrix multiplication : \n{multiply1}\n')

# Another way to do that is :
multiply2= mat1.dot(mat2)
print(f'Matrix multiplication : \n{multiply2}\n')

# Transpose of matrix
transpose1=mat1.transpose()
print(f'Transpose of matrix : \n{transpose1}\n')

# Or
transpose2=mat1.T
print(f'Transpose of matrix : \n{transpose2}\n')

# Inverse of matrix
inverse1=np.linalg.inv(mat1)
print(f'Inverse of matrix : \n{inverse1}\n')

```

#Spliting opeartion

```python
# Spliting
##> Splitting means to disjoint or break a an array in equal divisions  .
# For example:

splitting1=np.arange(1,10).reshape(3,3)
print(f'{splitting1}\n')

# 1. Horizontal split :
print(f'Horizontal split : \n{np.hsplit(splitting1,3)}\n')

# 2. Vertical split :
print(f'Vertical split : \n{np.vsplit(splitting1,3)}\n')

##> For better prsentaion :
for i in np.hsplit(splitting1,3):
  print(i , "\n")

for j in np.vsplit(splitting1,3):
  print(j,"\n")
```

```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],

    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],

    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
print(s)
```

```python
# Q2. Sum of each column ?

##> My approch .
q= np.sum(s,axis=0)
print(f'The sum of the column is : {q}\n')

for j in q:
  if j !=45:
    print("invaild sudoku")
    break
else:
  print("for columns the vaild sudoku")

##> This approch is correct .
```

#Student matrix

Q2. You are given two dimensional matrix of student data.


1.   Get the shape of the matrix.
2.   Find the average of age of the student.
3.   Extract Math marks of all the student .
4.   Find the highest marks scored in Science .
5.   Get the details of the student who scored more than 90 marks in Math .
6.   Increase Math marks of all students by 5.
7.   Find how many students are younger than 19.
8.   Calculate the average marks in each subject (column-wise mean).
9.   Get data of students who scored at least 80 in both subjects.
10.  Replace all Science marks < 75 with 0.

```python
# The given matrix :

# Columns: [Age, Math Marks, Science Marks]
data = np.array([
    [18, 85, 78],   # Student 1
    [19, 92, 88],   # Student 2
    [17, 76, 95],   # Student 3
    [18, 65, 70],   # Student 4
    [20, 90, 85]    # Student 5
])
```

```python
#1. Get the shape of the matrix.

##> My approch.
print(f'The shape of the matrix : {np.shape(data)}')

##> This is a vaild approch .
```

```python
# 2. Find the average of age of the student.

##> My approch .
print(f'The average of age of the student : {np.mean(data[:,0])}')

##> This is a correct approch.
```

```python
# 3. Extract Math marks of all the student .
print(f'Math marks of all the student : {data[:,1]}')

##> This is a correct approch.
```

```python
# 4. Find the highest marks scored in Science .

##> My Approch
print(f'Highest marks in Science : {np.max(data[:,2])}')

##> This is a correct approch .
```

```python
# 5. Get the details of the student who scored more than 90 marks in Math .

##> My Approch .
# Here we ca use boolean operation by giving a condition of marks > 90 while slicing the Math marks.
# And to pick that entire section we'll use the indexing .
# So if we combine the both ofthem we'll get :

print(f'Details of the student who scored more than 90 marks in Math : \n{data[data[:,1]>90]}')

##> This is a correct approch .
```

```python
# 6. Increase Math marks of all students by 5.
##> My approch .

# As we know that numpy allows broadcasting so if I extaract the math's marks by silcing.
# ad then add 5 to them the entire extracted element(math's marks) will be increased by 5.
# SO ,
print(f"Actual Math's marks : \n{data[:,1]}\n")
print(f"Math's Marks incresed by 5: \n{data[:,1]+5}")

##> This is a correct approch .
```

```python
# 7. Find how many students are younger than 19.
print(f' The students below 19 :\n{data[data[0:]<19]}')
```

```python
# 8. Calculate the average marks in each subject (column-wise mean).

print(f'Average marks in Math : {np.mean(data[:,1])}')
print(f"Average marks in Science : {np.mean(data[:,2])}")
```

```python
# 9. Get data of students who scored at least 80 in both subjects.

##> My approch
# In order to extract the data of the student who has scored more than 80 in both Math and Science.
# We need to to extract both Math & science marks columns and apply conditions on them .
# Also we have to pick the marks above 80 seperately form both the columns and check wheather they are true by bitwise 'and'.

print(f"student's data who scored at least 80 in both subjects : \n{data[data[:,1]>80] & data[data[:,2]>80] }")

##> This is a correct approch .
```
