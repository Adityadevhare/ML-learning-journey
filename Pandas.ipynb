#1. Pandas Series
###A series is a one-dimensional labeled array capable of holding any data type . The axis lables are collectively called the index .

```python
##> Crafting a series

import numpy as np
import pandas as pd

lables=['a','b','c']
my_list=[10,20,30]
arr = np.array([10,20,30])
d ={1:10, 2:20, 3:30}

print(f'Created series using list :\n{pd.Series(my_list)} \n')
print(f"Created series using numpy array :\n{pd.Series(arr)} \n")
print(f"Created series using dictionary :\n{pd.Series(d)} \n")

print(f"Labeling the series :\n{pd.Series(data=my_list, index=lables)} \n")
print(f"Labeling the series :\n{pd.Series(my_list, lables)} \n")

##? We can't create an series using 2-D numpy array because the Series itself is 1-D.
```

#2. Data Frames
###When we combine multiple series it creates a Dataframe .

####i. Creating a dataframe

```python
##> Creating Dataframe using Dictionary.
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago'],
    'Salary': [50000, 60000, 70000, 80000]
}

df=pd.DataFrame(data)
print(f'DataFrame using dictionary  :\n {df}\n\n')


##> Creating DataFrams using List.
data_list = [
    ["Alice" , 25 , "New york" , 50000],
    ["Bob" , 30 , "San Francisco" , 60000],
    ["Charlie" , 35 , "Los Angeles" , 70000],
    ["David" , 40 , "Chicago" , 80000]
]

df1=pd.DataFrame(data_list, columns = ["Name" , "Age" , "City" , "Salary"])
print(f'Dataframe using list :\n{df1}\n ')
```

###ii. Selection and Indexing of columns

```python
##> Indexing in Dataframes
# For example let us take the df1 dataframe to undersatnd indexing .

print(f'DataFrame : \n{df1}\n\n')

#> If I want to target the a specific index or columns of the dataframe.Then :
print(f'Targeting the Name column : \n{df1['Name']}\n\n')

#> But if I want to acess 2 or more columns at a same time then  I have to provide the list of list .
print(f'Targeting the Name and Agr of the Dataframe : \n{df1[['Name', 'Age']]}\n')
```

#iii. Creating a new column

```python
##> In oder to create or add a new column to an existing DataFrame ,
# Here I have already added a new column  to the existing Data Frame that is 'Designation'.

print(f'Original DataFrmae:\n{df1}\n')

# Adding a column
##> Here just like indexing we have to provide the name of the column which we want to add and then
#   provide the list of the elements "" , Imp: we have to provide the same no. of elements as in existing dataframe.
df1['Experience']=[2,3,5,6]

##> After adding a column
print(f'After change : \n{df1}\n')
```

#iv. Removing a column

```python
##> Let's take the existing DataFrame 'df1'.
# In order to remove the entire column we have to us the function 'drop'.

# Check if 'Experience' column exists before dropping and reassign the result
if 'Experience' in df1.columns:
    df1 = df1.drop('Experience', axis=1)
print(f'After removing the column : \n{df1}\n')
```

```python
df1
##> As you can see that the column that we removed is not really being removed ,
#   because drop creates a copy of the dataframe and shows the output.
#   To actully delet the column permenantly we have to give an implace value .

if 'Experience' in df1.columns:
    df1.drop('Experience', axis=1, inplace=True )
print(f'After removing the column : \n{df1}\n')
```
